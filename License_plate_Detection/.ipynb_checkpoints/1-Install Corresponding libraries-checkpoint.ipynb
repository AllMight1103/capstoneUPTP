# Install Tesseract

#### 1- Download Tesseract ocr from:
https://github.com/UB-Mannheim/tesseract/wiki


#### 2- Install Tesseract ocr

#### 3- After installing 
    - Open "Edit the system environment variables"
    - Go to "Environment Variables..."
    - On "System Variables", click on "path" and then press "Edit..."
    - Click on "NEW"
    - Add your tesseract.exe path. Ex: C:\Program Files\Tesseract-OCR\tesseract.exe


# Install cuda

### 1- Install CUDA Toolkit 11.8 from:
https://developer.nvidia.com/cuda-11-8-0-download-archive?target_os=Windows&target_arch=x86_64&target_version=11&target_type=exe_local


### 2- Download cudnn
https://developer.nvidia.com/rdp/cudnn-archive


### 3- Copying files to NVIDIA GPU Computing Toolkit
##### -Extract cudnn-windows-x86_64-8.9.5.30_cuda11-archive.zip
##### -Open the folder cudnn-windows-x86_64-8.9.5.30_cuda11-archive
##### -Go to bin and copy all files and paste it to "C:\Program Files\NVIDIA GPU Computing Toolkit\CUDA\v11.8\bin"
##### -Go to include and copy all files and paste it to "C:\Program Files\NVIDIA GPU Computing Toolkit\CUDA\v11.8\include"
##### -Go to lib->x64 and copy all files and paste it to "C:\Program Files\NVIDIA GPU Computing Toolkit\CUDA\v11.8\lib\x64"
